```python
import csv
f = open("guns.csv", "r")
reader = csv.reader(f)
data = list(reader)
print(data[:5])
```

```python
headers = data[0]
data = data[1:]
print(headers)
print(data[:5])
```

```python
years = []
for row in data:
    years.append(row[1])
    
year_counts = {}
for year in years:
    if year not in year_counts:
        year_counts[year] = 1
    else:
        year_counts[year] += 1

year_counts
```

```python
import datetime
dates = []
for row in data:
    year = int(row[1])
    month = int(row[2])
    date = datetime.datetime(year=year, month=month, day=1)
    dates.append(date)
dates[:5]
```

```python
date_counts = {}
for date in dates:
    if date not in date_counts:
        date_counts[date] = 1
    else:
        date_counts[date] += 1
date_counts
```

```python
sex_counts = {}
for row in data:
    sex = row[5]
    if sex not in sex_counts:
        sex_counts[sex] = 1
    else:
        sex_counts[sex] += 1
sex_counts
```

```python
race_counts = {}
for row in data:
    race = row[7]
    if race not in race_counts:
        race_counts[race] = 1
    else:
        race_counts[race] += 1
race_counts
```

*What we have learned so far*
In this dataset, far more men (86,349) than women (14,449) are killed by guns.  The largest racial group in the gun deaths dataset is White (66,237), followed by Black (23,296).  Hispanic is a distant third, at 9,022 deaths.  This may in part be a result of the overall ethnic makeup of the United States.

The month with the lowest gun violence seems to be February.  Across years, there is no visible major difference.
